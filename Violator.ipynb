```python
from ultralytics import YOLO
import cv2

MODEL_PATH = "/content/runs/detect/train/weights/best.pt"
VIDEO_PATH = "/content/tr.mp4"
OUTPUT_PATH = "/content/traffic_violation_output.mp4"

model = YOLO(MODEL_PATH)

STOP_LINE_Y = 430

CONF_THRES = 0.35
IMGSZ = 800
RED_CLASS = "red-light"
GREEN_CLASS = "green-light"
YELLOW_CLASS = "yellowlight"
VEHICLE_CLASS = "vehicles"


passed_on_red = set()
passed_on_green = set()
passed_on_yellow = set()
passed_on_unknown = set()

violators = set()

last_vehicle_centers = {}

vehicle_crossing_state = {}

last_light_state = "unknown"

def get_box_center_xyxy(x1, y1, x2, y2):
    cx = int((x1 + x2) / 2)
    cy = int((y1 + y2) / 2)
    return cx, cy


def get_light_state_from_detections(boxes, names):

    candidates = []

    for box in boxes:
        cls_id = int(box.cls[0])
        conf = float(box.conf[0])
        class_name = names[cls_id]

        if conf < CONF_THRES:
            continue

        if class_name == RED_CLASS:
            candidates.append(("red", conf))

        elif class_name == GREEN_CLASS:
            candidates.append(("green", conf))

        elif class_name == YELLOW_CLASS:
            candidates.append(("yellow", conf))

    if not candidates:
        return "unknown"

    candidates.sort(key=lambda x: x[1], reverse=True)
    return candidates[0][0]


def crossed_stop_line(prev_y, current_y, line_y):
    crossed_down = prev_y < line_y <= current_y
    crossed_up = prev_y > line_y >= current_y
    return crossed_down or crossed_up


def draw_info_panel(frame, light_state):
    h, w = frame.shape[:2]

    cv2.rectangle(frame, (20, 20), (420, 180), (0, 0, 0), -1)
    cv2.rectangle(frame, (20, 20), (420, 180), (255, 255, 255), 2)
    if light_state == "red":
        color = (0, 0, 255)
    elif light_state == "green":
        color = (0, 255, 0)
    elif light_state == "yellow":
        color = (0, 255, 255)
    else:
        color = (180, 180, 180)

    cv2.putText(frame, f"Light: {light_state.upper()}", (35, 55),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8, color, 2)

    cv2.putText(frame, f"Passed on RED: {len(passed_on_red)}", (35, 90),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 0, 255), 2)

    cv2.putText(frame, f"Passed on GREEN: {len(passed_on_green)}", (35, 120),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 0), 2)

    cv2.putText(frame, f"Passed on YELLOW: {len(passed_on_yellow)}", (35, 150),
                cv2.FONT_HERSHEY_SIMPLEX, 0.7, (0, 255, 255), 2)

    return frame

cap = cv2.VideoCapture(VIDEO_PATH)

width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = cap.get(cv2.CAP_PROP_FPS)

out = cv2.VideoWriter(
    OUTPUT_PATH,
    cv2.VideoWriter_fourcc(*"mp4v"),
    fps,
    (width, height)
)

while True:
    ret, frame = cap.read()

    if not ret:
        break
    results = model.track(
        frame,
        imgsz=IMGSZ,
        conf=CONF_THRES,
        persist=True,
        tracker="bytetrack.yaml",
        verbose=False
    )[0]

    boxes = results.boxes

    detected_light_state = get_light_state_from_detections(boxes, model.names)

    if detected_light_state != "unknown":
        last_light_state = detected_light_state

    light_state = last_light_state

    cv2.line(frame, (0, STOP_LINE_Y), (width, STOP_LINE_Y), (255, 255, 255), 3)
    cv2.putText(frame, "STOP LINE", (20, STOP_LINE_Y - 10),
                cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)
    for box in boxes:
        cls_id = int(box.cls[0])
        conf = float(box.conf[0])
        class_name = model.names[cls_id]

        if conf < CONF_THRES:
            continue

        x1, y1, x2, y2 = map(int, box.xyxy[0])

        if class_name != VEHICLE_CLASS:
            continue
        if box.id is None:
            continue

        track_id = int(box.id[0])
        cx, cy = get_box_center_xyxy(x1, y1, x2, y2)
        if track_id not in last_vehicle_centers:
            last_vehicle_centers[track_id] = (cx, cy)

        prev_cx, prev_cy = last_vehicle_centers[track_id]
        if track_id not in vehicle_crossing_state:
            if crossed_stop_line(prev_cy, cy, STOP_LINE_Y):

                vehicle_crossing_state[track_id] = light_state

                if light_state == "red":
                    passed_on_red.add(track_id)
                    violators.add(track_id)

                elif light_state == "green":
                    passed_on_green.add(track_id)

                elif light_state == "yellow":
                    passed_on_yellow.add(track_id)

                else:
                    passed_on_unknown.add(track_id)
        last_vehicle_centers[track_id] = (cx, cy)
        if track_id in violators:
            box_color = (0, 0, 255)
            label = f"ID {track_id} - VIOLATION"
        else:
            box_color = (0, 255, 0)
            label = f"ID {track_id}"

        cv2.rectangle(frame, (x1, y1), (x2, y2), box_color, 2)

        cv2.putText(frame, label, (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.7, box_color, 2)

        cv2.circle(frame, (cx, cy), 4, box_color, -1)
    frame = draw_info_panel(frame, light_state)

    out.write(frame)

cap.release()
out.release()

print("Done!")
print("Saved to:", OUTPUT_PATH)

print("================================")
print("Cars passed on RED:", sorted(list(passed_on_red)))
print("Cars passed on GREEN:", sorted(list(passed_on_green)))
print("Cars passed on YELLOW:", sorted(list(passed_on_yellow)))
print("Cars passed on UNKNOWN:", sorted(list(passed_on_unknown)))
print("Violators IDs:", sorted(list(violators)))
print("================================")
```
